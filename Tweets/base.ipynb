```python
import spacy
import nltk
from spacy import displacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import json
import pandas as pd
```

```python
stopWord = stopwords.words('french')
nlp = spacy.load('fr_core_news_md') 
```

```python
def recuperation(): 
    with open("./tweets/vpecresse.jsonl", "r") as r:
        tweets = r.read().split("\n")
        content = []
        bannedwords = ["#","https","@"]
        for tweet in tweets:
            try:
                data = json.loads(tweet)
                txt = data["content"].lower()
                txt_nlp = nlp(txt.lower())
                txtTraite = ""
                for token in txt_nlp:
                    if token.pos_ != "PUNCT" and token.lemma_ not in stopWord:
                        txtTraite += token.lemma_ + " "
                new_text = txtTraite.removesuffix(" ")
                content.append(new_text)
            except:
                pass
            # break
    return content
```

```python
word = recuperation()
word
```

```python
fdist = FreqDist()
for i in range(len(word)):
    for j in word_tokenize(word[i]):
        fdist[j] += 1
    
#fdist
fdist.most_common(10)
```
