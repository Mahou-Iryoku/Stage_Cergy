# Exemple de scraping des sites de campagnes

## Premier objectif

Extraire les programmes des différents sites.

NOM CANDIDAT | LIEN | TITRE | DESCRIPTION

## Librairies pour effectuer du web scrapping

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
opts = Options()
opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36")
```

## Ouverture du navigateur Google Chrome

```python
browser = webdriver.Chrome(ChromeDriverManager().install(), chrome_options=opts)
```

## Phase extraction de données des candidats

```python
def extractionssitescandidats(name,url,selector_title,selector_url,selector_description):
    tab_links = []
    candidat = []
    if url:
        browser.get(url)
        if selector_title or selection_description:
            titres = browser.find_elements(By.CSS_SELECTOR,selector_title)
            description = browser.find_elements(By.CSS_SELECTOR,selector_description)
            if selector_url:
                for element in titres:
                    link = element.find_element(By.CSS_SELECTOR,selector_url).get_attribute("href")
                    tab_links.append(link)
                    
                if selector_description:
                    for i in range(len(titres)):
                        data = {}
                        data['Nom du candidat'] = name
                        data['URL'] =  tab_links[i]
                        data['Titre du programme'] = titres[i].text
                        data['Description'] = description[i].text
                        candidat.append(data)
                else: 
                    print("Les descriptions n'existent pas ",selector_description)
                    None

            else:
                print("Les liens du programme n'existent pas",selector_url)
                None
        else:
            print("Les titres du programme n'existent pas",selector_title)
            None
    else:
        print("Le lien du candidat n'existe pas")
        None
        
    df = pd.DataFrame(candidat)
    return df
```
